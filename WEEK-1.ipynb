# Functions, Import Statements:

```python
import numpy as np
import matplotlib.pyplot as plt
def Mean(data):
    s=0
    n=0
    for i in data:
        s+=i  
        n+=1
    return s/n

def count(data):
    c=0
    for i in data:
        c+=1
    return c

def Sum(data):
    s=0
    for i in data:
        s+=i
    return s
```

# Regression Line:

```python
import matplotlib.pyplot as plt
def reg(x,y):
    c=x*y
    sigxy=Sum(c)
    sigx=Sum(x)
    sigy=Sum(y)
    x2=x*x
    sigx2=Sum(x2)
    n=count(y)
    m=((n*sigxy)-(sigx*sigy))/((n*sigx2)-(sigx**2))
    c=(sigy-(m*sigx))/n
    ycap=m*x+c
    ybar=Mean(y)
    SST=0
    for i in range(n):
        SST+=(y[i]-ybar)**2
    SSE=0
    for i in range(n):
        SSE+=(y[i]-ycap[i])**2
    
    plt.plot(x,y)
    plt.plot(x,ycap)
    plt.xlabel('x - axis')
    plt.ylabel('y - axis')
    plt.title('Linear Regression and Best Fit')
    plt.show()
    
    
   # plt.xlabel('x - axis')
   # plt.ylabel('y - axis')
    #plt.title('Best Fit')
    #plt.show()
        
    r=1-(SSE/SST)
    if(r>0.9):
        print('Best Fit')
    else:
        print('Not a Best Fit')
        
x=list(map(float,input('Enter x values:').split()))
x=np.array(x)
y=list(map(float,input('Enter y values:').split()))
y=np.array(y)
reg(x,y)
```
